```python
import numpy as np
from mogutda import SimplicialComplex
import networkx as nx
import matplotlib.pyplot as plt
```

Torus (simplicial complex)

imagetorussimplcompl.png


```python
torus_sc = [(7,3,1), (1,2,3), (2,5,7), (2,3,6), (2,5,6), (5,7,3), (3,4,6), (6,5,1), (1,5,4), (3,4,5), (7,4,6), (7,6,1), (1,4,2), (2,4,7)]
torus_c = SimplicialComplex(simplices=torus_sc)
```

```python
from IPython.display import display, Markdown

display(Markdown(
   rf"""
$\beta_0 ={torus_c.betti_number(0)}$

$\beta_1 ={torus_c.betti_number(1)}$

$\beta_2 ={torus_c.betti_number(2)}$


"""))
#print(torus_c.betti_number(0))   
#print(torus_c.betti_number(1))  
#print(torus_c.betti_number(2)) 
n1 = nx.Graph()
n1.add_edges_from(torus_c.n_faces(1))
nx.draw(n1)
plt.show()  
```
