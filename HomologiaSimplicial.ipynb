Simplicial Complex Library 

```python
import numpy as np
from mogutda import SimplicialComplex
import networkx as nx
import matplotlib.pyplot as plt
```

Torus (simplicial complex)

imagetorussimplcompl.png


```python
torus_sc = [(7,3,1), (1,2,3), (2,5,7), (2,3,6), (2,5,6), (5,7,3), (3,4,6), (6,5,1), (1,5,4), (3,4,5), (7,4,6), (7,6,1), (1,4,2), (2,4,7)]
torus_c = SimplicialComplex(simplices=torus_sc)
```

Calculate Betti numbers

```python
from IPython.display import display, Markdown

display(Markdown(
   rf"""
$\beta_0 ={torus_c.betti_number(0)}$

$\beta_1 ={torus_c.betti_number(1)}$

$\beta_2 ={torus_c.betti_number(2)}$


"""))
#print(torus_c.betti_number(0))   
#print(torus_c.betti_number(1))  
#print(torus_c.betti_number(2)) 
n1 = nx.Graph()
n1.add_edges_from(torus_c.n_faces(1))
nx.draw(n1)
plt.show()  
```

Get the Betti numbers of the cylinder
$$\{(x,y,z)\vert \:x^2+y^2=1\text{ and } z\in\left[0,1\right]\}$$

Captura de pantalla 2023-04-11 a la(s) 23.22.20.png

```python
cylinder_sc = [(1,2,4), (1,3,4), (3,4,6), (3,5,6), (5,6,2), (5,2,1)]
cylinder_c = SimplicialComplex(simplices=cylinder_sc)
n2 = nx.Graph()
n2.add_edges_from(cylinder_c.n_faces(1))
nx.draw(n2)
plt.show() 
```

```python
display(Markdown(
   rf"""
$\beta_0 ={cylinder_c.betti_number(0)}$

$\beta_1 ={cylinder_c.betti_number(1)}$

$\beta_2 ={cylinder_c.betti_number(2)}$


"""))

```

Cone $$\{(x,y,z)\vert x^2+y^2=z^2\text{ and } z\in\left[0,1\right]\}$$

```python
cone_sc = [(1,2,3), (1,3,4), (1,4,2)]
cone_c = SimplicialComplex(simplices=cone_sc)

n4 = nx.Graph()
n4.add_edges_from(cone_c.n_faces(1))
nx.draw(n4)
plt.show() 

display(Markdown(
   rf"""
$\beta_0 ={cone_c.betti_number(0)}$

$\beta_1 ={cone_c.betti_number(1)}$

$\beta_2 ={cone_c.betti_number(2)}$


"""))
```

```python
sphere_sc = [(1,2,3), (1,3,4), (1,4,2), (2,3,4)]
sphere_c = SimplicialComplex(simplices=sphere_sc)
display(Markdown(
   rf"""
$\beta_0 ={sphere_c.betti_number(0)}$

$\beta_1 ={sphere_c.betti_number(1)}$

$\beta_2 ={sphere_c.betti_number(2)}$


"""))
```
